```python
import sys
import os 

import pandas as pd
import numpy as np  

from bs4 import BeautifulSoup  
from selenium import webdriver  
from selenium.webdriver.common.keys import Keys
import chromedriver_autoinstaller

import time   
from tqdm import tqdm_notebook 

from selenium.webdriver.chrome.options import Options

import warnings
warnings.filterwarnings('ignore')

number_list = []
name_list = []

# pip install chromedriver_autoinstaller (없으면 깔아야함)
chrome_path = chromedriver_autoinstaller.install()


# Set up the options for headless browsing
chrome_options = Options()
# chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--no-sandbox')

# Create a new instance of the Chrome driver with the options set
driver = webdriver.Chrome(chrome_path, chrome_options=chrome_options)

#season_set = [19]
#number_set = [1]

enternumber = [12, 18, 18, 16, 15]

#https://www.kovo.co.kr/game/v-league/11141_game-summary.asp?season=019&g_part=201&r_round=4&g_num=1&

for season in range(19, 20):
    print(season)
    for number in range(1, 2):
        sitename = f'https://www.kovo.co.kr/game/v-league/11141_game-summary.asp?season={season:03d}&g_part=201&r_round=1&g_num={number:d}&'

        print(sitename)

        driver.get(sitename)

        classname = "lst_board.lst_fixed.w123"
        table = driver.find_element_by_class_name(classname)

        # find the tbody element within the parent element
        thead_element = table.find_element_by_tag_name("thead")

        # find the tbody element within the parent element
        tbody_element = table.find_element_by_tag_name("tbody")

        # rows = tbody_element.find_elements_by_tag_name('td')

        rows = tbody_element.find_elements_by_tag_name('td')

        print(f'name')
        for ik, row in enumerate(rows):
            text = row.get_attribute('innerHTML')

            if '>'in text:
                text = text[text.find('>')+1:]
                text = text[:text.find('<')]
            
            print(text + ' ', end='')
            if ik % 2 == 1:
                print('')
                number_list.append(ik)
                name_list.append(text)

        print('')
        for i in range(5):
            globals()["data_list{}".format(i)] = []
            print(f'page number {i+1}')
            classname = f'lst_board.lst_scroll.w837.record_{i+1}'
            
            table = driver.find_element_by_class_name(classname)

            # find the tbody element within the parent element
            thead_element = table.find_element_by_tag_name("thead")


            # find the tbody element within the parent element
            tbody_element = table.find_element_by_tag_name("tbody")


            rows = tbody_element.find_elements_by_tag_name('td')
            
            data_temp = []

            for ik, row in enumerate(rows):
                text = row.get_attribute('innerHTML')
                    
                if '>'in text:
                    text = text[text.find('>')+1:]
                    text = text[:text.find('<')]
                print(text + ' ', end='')

                data_temp.append(text)
                
                if ik % enternumber[i] == enternumber[i]-1:
                    print('줄바꿈')
                    globals()["data_list{}".format(i)].append(data_temp)
                    # print(data_temp)
                    data_temp = []
            print('')



# Close the driver
driver.quit()
```

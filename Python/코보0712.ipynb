```python
import sys
import os
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import chromedriver_autoinstaller
import time
from tqdm import tqdm_notebook
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import warnings

warnings.filterwarnings('ignore')

number_list = []
name_list = []

chrome_path = chromedriver_autoinstaller.install()
chrome_options = Options()
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--no-sandbox')
driver = webdriver.Chrome(chrome_path, chrome_options=chrome_options)

wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds

enternumber = [12, 18, 18, 16, 15]

for season in range(19, 20):
    print(season)
    for number in range(1, 2):
        sitename = f'https://www.kovo.co.kr/game/v-league/11141_game-summary.asp?season={season:03d}&g_part=201&r_round=1&g_num={number:d}&'

        print(sitename)

        driver.get(sitename)

        # Process each div element using the provided CSS selectors
        css_selectors = ["#team1 > div > div.wrp_lst > table.lst_board.lst_fixed.w123",
                         "#team1 > div > div.wrp_lst > table.lst_board.lst_scroll.w837.record_1",
                         "#team1 > div > div.wrp_lst > table.lst_board.lst_scroll.w837.record_2",
                         "#team1 > div > div.wrp_lst > table.lst_board.lst_scroll.w837.record_3",
                         "#team1 > div > div.wrp_lst > table.lst_board.lst_scroll.w837.record_4",
                         "#team1 > div > div.wrp_lst > table.lst_board.lst_scroll.w837.record_5"
                        ]

        for selector in css_selectors:
            div_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, selector)))
            div_inner_html = div_element.get_attribute('innerHTML')
            print(div_inner_html)

        print('---')

# Close the driver
driver.quit()
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.kovo.co.kr/game/v-league/11141_game-summary.asp?season=019&g_part=201&r_round=1&g_num=1"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find the table containing the player data
table = soup.find("table", {"summary": "선수별 경기일정 상세안내 표로 No, 로 구성"})

# Create empty lists to store the player numbers and names
numbers = []
names = []

# Iterate over the rows in the table body
for row in table.find("tbody").find_all("tr"):
    # Find the cells in each row
    cells = row.find_all("td")

    # Extract the player number and name from the cells
    number = cells[0].text.strip()
    name = cells[1].text.strip().split()[0]  # Extract only the player's name (excluding position)

    # Append the data to the respective lists
    numbers.append(number)
    names.append(name)

# Print the player numbers and names
for number, name in zip(numbers, names):
    print(f"No: {number}, Name: {name}")
```

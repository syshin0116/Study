## 2_1. 아래 결과 화면과 같이 원하는 메뉴를 입력받아 해당 메뉴를 수행하는 프로그램을 작성하시오.

**1. 두 수 사이의 홀수 합 구하기**  
: 두 수를 입력받아 두 수 사이의 홀수를 한줄에 출력하고,  합도 출력하시오.  
**2. 구구단**  
:  원하는 단을 입력받아 곱하는 수는 아래와 같이 3, 6, 9만 출력되도록 하시오.(반드시 반복문 사용하여 출력)  
  
    
**[조건]** 
* 프로그램 시작 첫 줄에 [본인학과학번이름] 출력하시오. 
* 메뉴입력 시, 1, 2 외 다른 것 입력 할때 '입력오류!! 1 또는 2만 입력하세요' 라고 출력하시오.
* 'n'를 입력할 때까지 계속 반복 수행되도록 하시오.  
  - y를 입력 시 계속 반복  
  - n를 입력 시 '프로그램을  종료합니다.' 출력하고 종료
  
**[결과화면]**  

그림1.png
 


```python
print('[본인학번이름작성]')
print('아래 메뉴에서 원하여 번호를 입력하세요')
while True:
    print('[*** 메뉴 ***]')
    menu = input('1.두 수 사이의 홀수 합 구하기  2.구구단 :')
    if menu == '1':
        num1 = int(input('** 첫 번째 숫자를 입력하세요:'))
        num2 = int(input('** 두 번째 숫자를 입력하세요:'))
        total = 0
        for i in range(num1, num2+1):
            if i % 2 == 1:
                print(i, end=' ')
                total += i
        print(f'\n{num1}과 {num2}사이의 홀수 합 : {total}')
    elif menu == '2':
        num3 = int(input('원하는 단을 입력하세요:'))
        for i in range(3, 10, 3):
            print(f'{num3} X {i} = {num3*i}')
    else:
        print('입력오류!! 1 또는 2만 입력하세요')
    menu2 = input('계속하시겟습니까?(y/n):')
    if menu2 == 'n':
        print('프로그램을 종료합니다.!')
        break
    elif menu2 == 'y':
        continue
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = (
    "Apple",
    "Grapes",
    "Melon",
    "Banana",
)
sizes = [34, 18, 16, 32]
colors = ["blue", "red", "green", "orange"]

# Plot
plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    startangle=140,
)

plt.axis("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt

pie_labels = ("Apple", "Grapes", "Melon", "Banana")
pie_sizes = [34, 18, 16, 32]
pie_colors = ["blue", "red", "green", "orange"]

bar_labels = ["a", "b", "c", "d"]
bar_values = [3, 7, 4, 9]
bar_colors = ["black", "red", "blue", "cyan"]

plt.rcParams.update({"font.size": 14})

fig, (ax1, ax2) = plt.subplots(
    1, 2, figsize=(16, 8), gridspec_kw={"width_ratios": [3, 6]}
)

ax1.pie(
    pie_sizes,
    labels=pie_labels,
    colors=pie_colors,
    autopct="%1.1f%%",
    startangle=140,
)
ax1.set_anchor("S")

ax2.bar(bar_labels, bar_values, color=bar_colors)
ax2.set_title("Vertical Bar Graph", fontsize=20)
ax2.set_yticks(range(0, max(bar_values), 2))

plt.show()
```

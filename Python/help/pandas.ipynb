```python
month = int(input("월 입력:"))
if 3 <= month <= 5:
    season = '봄'
elif 6 <= month <= 8:
    season = '여름'
elif 9 <= month <= 11:
    season = '가을'
else:
    season = '겨울'
print('{}월은 {}'.format(month, season))
```

```python
class Node:
    def __init__(self,item):
        self.data = item
        self.next = None

class LStack:
    def __init__(self):
        self.head = None  
        self.tail = None
        self.size=0

    def isEmpty(self):
        return not self.head

    def push(self,item):  #마지막에 넣기
        new_node = Node(item)
        
        if self.isEmpty(): #노드가 없다면
            self.head = new_node
            self.tail = new_node
        else:             #무언가가 있다면
            self.tail.next = new_node
            self.tail = new_node
        self.size +=1
            
    def pop(self):
        
        if self.isEmpty():
            return

        else:
            if self.size == 1: #노드가 1개
                value = self.tail.data
                self.head = None
                self.tail = None
                self.size -=1
                return value
                
            else: #노드가 2개 이상
                value = self.tail.data
                current_node = self.head
                
                for i in range(self.size):
                    if current_node.next is self.tail:
                        self.tail = current_node
                        break
                    
                    current_node = current_node.next
                current_node.next = None    
                self.size -=1
                return value


    def top(self):

        if self.isEmpty():
            return
        else:
            value = self.tail.data
            return value 

def checkparenthesis(expr):
    S = LStack()
    for i in expr:
        if i == "(":
            S.push(i)
        elif i ==")":
            if S.isEmpty():
                return False
            else:
                pop_result = S.pop()
                if pop_result!="(":
                    return False

    return S.isEmpty()

print(checkparenthesis("(((())))"))
print(checkparenthesis("()()"))
print(checkparenthesis("((()))"))
print(checkparenthesis("(()))"))
```

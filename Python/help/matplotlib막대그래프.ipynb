```python
import pandas as pd
import matplotlib.pyplot as plt

first_child = pd.DataFrame({
    '항목': ['육아 관련 온라인까페', '유명 영유아부모 블로그, SNS', '주변지인, 모임, 지인 SNS 등', 'TV/인터넷', '육아업체 광고, 홈페이지', '정부 및 공공기관 홈페이지 및 인쇄물'],
    '정보를 가장 많이 얻는 채널': [76, 27, 58, 34, 2, 3],
    '신뢰하는 채널': [59, 23, 60, 37, 0, 21]
})

fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
bar1 = ax.barh(first_child['항목'], first_child['정보를 가장 많이 얻는 채널'], bar_width, label='정보를 가장 많이 얻는 채널')
bar2 = ax.barh(first_child['항목'], first_child['신뢰하는 채널'], bar_width, left=first_child['정보를 가장 많이 얻는 채널'], label='신뢰하는 채널')

total1 = sum(first_child['정보를 가장 많이 얻는 채널'])
total2 = sum(first_child['신뢰하는 채널'])
for b1, b2 in zip(bar1, bar2):
    width1 = b1.get_width()
    width2 = b2.get_width()
    percentage1 = width1 / total1 * 100
    percentage2 = width2 / total2 * 100
    ax.text(width1 + 2, b1.get_y() + b1.get_height() / 2, f'{width1}\n{percentage1:.1f}%', ha='left', va='center')
    ax.text(width1 + width2 + 2, b2.get_y() + b2.get_height() / 2, f'{width2}\n{percentage2:.1f}%', ha='left', va='center')

ax.set_xlabel('응답 수')
ax.set_ylabel('채널')
ax.set_title('정보를 가장 많이 얻는 채널 vs. 신뢰하는 채널')

ax.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#차트글꼴
plt.rc('font', family='Malgun Gothic')


first_child = pd.DataFrame({
    '항목': ['육아 관련 온라인까페', '유명 영유아부모 블로그, SNS', '주변지인, 모임, 지인 SNS 등', 'TV/인터넷', '육아업체 광고, 홈페이지', '정부 및 공공기관 홈페이지 및 인쇄물'],
    '정보를 가장 많이 얻는 채널': [76, 27, 58, 34, 2, 3],
    '신뢰하는 채널': [59, 23, 60, 37, 0, 21]
})

second_child = pd.DataFrame({
    '항목': ['육아 관련 온라인까페', '유명 영유아부모 블로그, SNS', '주변지인, 모임, 지인 SNS 등', 'TV/인터넷', '육아업체 광고, 홈페이지', '정부 및 공공기관 홈페이지 및 인쇄물'],
    '정보를 가장 많이 얻는 채널': [25, 11, 38, 223, 1, 0],
    '신뢰하는 채널': [22, 13, 33, 21, 1, 8]
})

first_child.to_csv('first_child.csv')
second_child.to_csv('second_child.csv')

labels = first_child['항목']
x = np.arange(len(labels))
width = 0.35

fig, (ax1, ax2) = plt.subplots(1, 2,  figsize=(20, 10))

rects1 = ax1.barh(x - width/2, first_child['정보를 가장 많이 얻는 채널'], width, label='정보를 가장 많이 얻는 채널')
rects2 = ax1.barh(x + width/2, first_child['신뢰하는 채널'], width, label='신뢰하는 채널')

for rect in rects1:
    width = rect.get_width()
    percentage = width / sum(first_child['정보를 가장 많이 얻는 채널']) * 100
    ax1.annotate(f'{width}\n{percentage:.1f}%', xy=(width, rect.get_y() + rect.get_height() / 2),
                xytext=(3, 0), textcoords="offset points", ha='left', va='center')
    
for rect in rects2:
    width = rect.get_width()
    percentage = width / sum(first_child['신뢰하는 채널']) * 100
    ax1.annotate(f'{width}\n{percentage:.1f}%', xy=(width, rect.get_y() + rect.get_height() / 2),
                xytext=(3, 0), textcoords="offset points", ha='left', va='center')

ax1.set_yticks(x)
ax1.set_yticklabels(labels, multialignment='center')
ax1.set_xlabel('응답 수')
ax1.set_ylabel('채널')
ax1.set_title('첫째자녀 기준')

labels = second_child['항목']
x = np.arange(len(labels))
width = 0.35

rects1 = ax2.barh(x - width/2, second_child['정보를 가장 많이 얻는 채널'], width, label='정보를 가장 많이 얻는 채널')
# 두 번째 그래프 그리기
rects2 = ax2.barh(x + width/2, second_child['신뢰하는 채널'], width, label='신뢰하는 채널')

for rect in rects1:
    width = rect.get_width()
    percentage = width / sum(second_child['정보를 가장 많이 얻는 채널']) * 100
    ax2.annotate(f'{width}\n{percentage:.1f}%', xy=(width, rect.get_y() + rect.get_height() / 2),
                xytext=(3, 0), textcoords="offset points", ha='left', va='center')
    
for rect in rects2:
    width = rect.get_width()
    percentage = width / sum(second_child['신뢰하는 채널']) * 100
    ax2.annotate(f'{width}\n{percentage:.1f}%', xy=(width, rect.get_y() + rect.get_height() / 2),
                xytext=(3, 0), textcoords="offset points", ha='left', va='center')

ax2.set_yticks(x)
ax2.set_yticklabels(labels, multialignment='center')
ax2.set_xlabel('응답 수')
ax2.set_ylabel('채널')
ax2.set_title('둘째자녀 기준')


plt.legend(loc='upper center', bbox_to_anchor=(1.5, 1))
plt.subplots_adjust(wspace = 0.8)
plt.suptitle('육아 정보를 얻는 경로')
plt.show()
```

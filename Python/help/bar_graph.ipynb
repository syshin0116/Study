```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('교통사고통계.csv', encoding='cp949')

plt.figure(figsize=(30, 10))

plt.bar(df.연도, df.사망, color='green')

plt.plot(df.연도, df.사망, marker='o', color='blue')

plt.xlabel('Year')
plt.ylabel('Death roll')
plt.xticks(df.연도)

plt.show()
```

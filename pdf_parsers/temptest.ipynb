```python
def replace_base64_encoding(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if key == "base64_encoding":
                data[key] = ""  # 값을 ''로 대체
            else:
                replace_base64_encoding(value)  # 재귀적으로 탐색
    elif isinstance(data, list):
        for item in data:
            replace_base64_encoding(item)  # 리스트 내부도 탐색
```

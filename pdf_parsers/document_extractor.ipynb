```python
INPUT_FILE_PATH = "./data/sample_pdf copy.pdf"
```

```python
from dotenv import load_dotenv, find_dotenv
import os
load_dotenv(find_dotenv())
```

```python
import requests

url = "https://api.va.landing.ai/v1/tools/agentic-document-analysis"
files = {
    "pdf": open(INPUT_FILE_PATH, "rb")
    # OR, for PDF
    # "pdf": open("{{path_to_file}}", "rb")
}
headers = {
    "Authorization": f"Basic {os.getenv('LANGDINGAI_API_KEY')}",
}
response = requests.post(url, files=files, headers=headers)

print(response.json())
```

```python
print(response.status_code)  # HTTP 상태 코드 출력
print(response)
```

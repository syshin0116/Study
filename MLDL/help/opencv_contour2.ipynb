```python
import cv2
import numpy as np
import argparse
import imutils
from imutils import perspective
from imutils import contours
from scipy.spatial import distance as dist


src = cv2.imread("puma.png", cv2.IMREAD_COLOR)
src = cv2.resize(src, (1024, 512))
dst = src.copy()
gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(gray, 75, 255, cv2.THRESH_OTSU)
gray = cv2.GaussianBlur(gray, (7,7), 0)
binary = cv2.bitwise_not(binary)
contours, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_TC89_L1)

#윤곽선
for i in range(len(contours)):
    cv2.drawContours(src, [contours[i]], 0, (0, 0, 255), 3)
    cv2.putText(src, str(i), tuple(contours[i][0][0]), cv2.FONT_HERSHEY_COMPLEX, 0, (0, 255, 0), 0)

    # 서클의 중심 좌표를 계산합니다.
    x = int(contours[i][4][0][0])  # x 좌표는 contours[i]의 중간 부분인 4번째 좌표를 사용합니다.
    y = int(contours[i][4][0][1])  # y 좌표도 마찬가지로 contours[i]의 중간 부분인 4번째 좌표를 사용합니다.

    circle1_x, circle1_y = x - 60, y + 5
    circle2_x, circle2_y = x - 60, y + 370
   
    circle3_x, circle3_y = x + 10, y - 5
    circle4_x, circle4_y = x - 130, y - 5
    
    circle5_x, circle5_y = x - 480, y + 65
    circle6_x, circle6_y = x + 355, y + 65
    
    circle7_x, circle7_y = x + 300, y + 140
    circle8_x, circle8_y = x - 425, y + 140
    
    circle9_x, circle9_y = x - 320, y + 155
    circle10_x, circle10_y = x + 200, y + 155

    # 서클을 새로운 좌표로 이동시킵니다.
    cv2.circle(src, (circle1_x, circle1_y), 3, (255, 255, 0), 3) 
    cv2.circle(src, (circle2_x, circle2_y), 3, (255, 255, 0), 3) 
    
    cv2.circle(src, (circle3_x, circle3_y), 3, (255, 0, 255), 3)  
    cv2.circle(src, (circle4_x, circle4_y), 3, (255, 0, 255), 3)  
    
    cv2.circle(src, (circle5_x, circle5_y), 3, (0, 0, 0), 3)  
    cv2.circle(src, (circle6_x, circle6_y), 3, (0, 0, 0), 3)  
    
    cv2.circle(src, (circle7_x, circle7_y), 3, (111, 111, 111), 3)
    cv2.circle(src, (circle8_x, circle8_y), 3, (111, 111, 111), 3)
   
    cv2.circle(src, (circle9_x, circle9_y), 3, (50, 50, 255), 3)
    cv2.circle(src, (circle10_x, circle10_y), 3, (50, 50, 255), 3)   
       
    
    # 서클 1과 서클 2 사이의 거리 계산
d1 = dist.euclidean((circle1_x, circle1_y), (circle2_x, circle2_y))

# 서클 3과 서클 4 사이의 거리 계산
d2 = dist.euclidean((circle3_x, circle3_y), (circle4_x, circle4_y))

# 서클 5과 서클 6 사이의 거리 계산
d3 = dist.euclidean((circle5_x, circle5_y), (circle6_x, circle6_y))

# 서클 7과 서클 8 사이의 거리 계산
d4 = dist.euclidean((circle7_x, circle7_y), (circle8_x, circle8_y))

# 서클 9과 서클 10 사이의 거리 계산
d5 = dist.euclidean((circle9_x, circle9_y), (circle10_x, circle10_y))


cv2.line(src, (circle1_x, circle1_y), (circle2_x, circle2_y), (255, 255, 0), 1)
cv2.line(src, (circle3_x, circle3_y), (circle4_x, circle4_y), (255, 0, 255), 1)
cv2.line(src, (circle5_x, circle5_y), (circle6_x, circle6_y), (0, 0, 0), 1)
cv2.line(src, (circle7_x, circle7_y), (circle8_x, circle8_y), (111, 111, 111), 1)
cv2.line(src, (circle9_x, circle9_y), (circle10_x, circle10_y), (50, 50, 255), 1)

cv2.putText(src, "{:.2f} cm".format(d1),
(int(x), int(y+20)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (12, 255, 255), 1)

cv2.putText(src, "{:.2f} cm".format(d2),
(int(x), int(y+40)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (12, 255, 255), 1)

cv2.putText(src, "{:.2f} cm".format(d3),
(int(x), int(y+60)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (12, 255, 255), 1)

cv2.putText(src, "{:.2f} cm".format(d4),
(int(x ), int(y+80)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (12, 255, 255), 1)

cv2.putText(src, "{:.2f} cm".format(d5),
(int(x), int(y+100)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (12, 255, 255), 1)

print(i, hierarchy[0][i])
cv2.imshow("src", src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# check_df에서 column명을 문제 상황에 맞게 수정하기 위하여 #####를 채워봅시다.
df = pd.read_html('http://ncov.mohw.go.kr/en/bdBoardList.do?brdId=16&brdGubun=163&dataGubun=&ncvContSeq=&contSeq=&board_id=')
check_df.filter = df[1].iloc[ :, :3]
check_df.columns = ["지역", "국가", "확진자(사망자)"]

# 출력을 통해 확인해봅시다.
check_df
```

```python

# check_df를 '국가'로 정렬해 check_df_sorted에 담으려고 합니다. #####를 채워봅시다.
# 힌트 : sort_values 함수 사용
check_df.sort_values('국가', ascending=False)
```

```python
# 제공하는 자료 중 3번째 자료에 해당하는 것은 사이트에서 확인할 수 있습니다.
# 출력하여 확인해봅니다.
check_df = df[2]
check_df
```

```python
# 동영상 수업에서 다루지 않은 내용이나 참고하시기 바랍니다!
# 위의 내용에서 NaN(Not A Number)이 계속 나타나고 있습니다. 
# 의미 없는 행(row)이므로 삭제해보도록합니다.
# NaN의 행을 삭제하는 경우 dropna()를 사용하시면 됩니다.
new_df = check_df.dropna()
new_df
```

```python
# 특정 국가의 자료만을 확인하려고 합니다.
# 나라 이름을 입력 받아 해당 국가의 자료만 출력해보세요.
nation = input ('국가명을 입력하세요 : ')
check_df[check_df['Country', 'Country']==nation]
```

```python
# 이중 컬럼명이므로 아래은 컬럼명으로 수정하기 위하여 #####를 완성하세요.
new_df.columns = ['국가', '확진자', '발생빈도', '사망자', '발생빈도', '치사율']
new_df
```

```python
# 입력 받은 국가명이 포함된 국가의 자료를 확인하기 위하여 #####를 완성하세요.
select = new_df['국가']==nation
pick = new_df[select]
pick
```

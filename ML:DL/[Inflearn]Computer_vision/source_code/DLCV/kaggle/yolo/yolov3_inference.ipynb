### 다른 image로 Detect 수행

```python
!mkdir /kaggle/working/data
!wget -O /kaggle/working/data/beatles01.jpg https://raw.githubusercontent.com/chulminkw/DLCV/master/data/image/beatles01.jpg
```

```python
# 실행 디렉토리를 고정하고, 시각화시 bounding box line 두께를 조절. 
!cd yolov3;python detect.py --weights yolov3.pt --img 640 --conf 0.25 --source /kaggle/working/data/beatles01.jpg  \
                            --project /kaggle/working/data  --name=run_image --exist-ok --line-thickness 1
```

```python
Image(filename='/kaggle/working/data/run_image/beatles01.jpg', width=600)
```

### Video Detection
* detect.py의 source가 mp4이면 자동으로 Video Detection 수행. 

```python
# 영상 다운로드 하기
!wget -O /kaggle/working/data/Night_Day_Chase.mp4 https://github.com/chulminkw/DLCV/blob/master/data/video/Night_Day_Chase.mp4?raw=true
```

```python
# kaggle kernel은 --project를 /kaggle/working/run_video 로 설정하여 Detect된 영상을 저장. 
!cd yolov3;python detect.py --weights yolov3.pt --img 640 --conf 0.25 --source /kaggle/working/data/Night_Day_Chase.mp4 \
                            --project=/kaggle/working/run_video --exist-ok --line-thickness 1
```

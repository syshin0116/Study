```python
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import xgboost as xgb  # xgboost 불러오기
from sklearn.model_selection import (
    train_test_split,
)  # 학습 및 테스트 데이터 분류함수 불러오기

import os

for dirname, _, filenames in os.walk("/kaggle/input"):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# 오리지날 데이터 활용
url = (
    "https://raw.githubusercontent.com/erkansirin78/datasets/master/Churn_Modelling.csv"
)
origin = pd.read_csv(url)
# 학습, 테스트, 제출 파일 읽기
train = pd.read_csv("/kaggle/input/knou-deeplearning2025-finalterm/train.csv")
test = pd.read_csv("/kaggle/input/knou-deeplearning2025-finalterm/test.csv")
sample_submission = pd.read_csv(
    "/kaggle/input/knou-deeplearning2025-finalterm/sample_submission.csv"
)
```

```python
# 오리지날 데이터 활용 학습데이터 통합
train_drop = train.drop(["id"], axis=1)
origin_drop = origin.drop(["RowNumber"], axis=1)
combined = pd.concat([train_drop, origin_drop], axis=0)

# 문자 데이터 및 결과('Exited') 제거 (팁: 제거가 아닌 변환을 통해 성능 개선 가능)
X = combined.drop(["Surname", "Geography", "Gender", "Exited"], axis=1)
y = combined.Exited

# 학습 및 테스트 데이터 분류
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=22
)
```

```python
# XGBoost 기반 학습

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

model = xgb.train(
    {
        "objective": "binary:logistic",  # 이진분류 셋팅
        "eval_metric": "logloss",  # 평가함수
        "max_depth": 1,  # 최대 트리 개수
    },
    dtrain,
    num_boost_round=10,
)

# 문자 데이터 제거 (팁: 제거가 아닌 변환을 통해 성능 개선 가능)
test_drop = test.drop(["id", "Surname", "Geography", "Gender"], axis=1)

test_drop = xgb.DMatrix(test_drop)
y_pred = model.predict(test_drop)
```

```python
sample_submission["Exited"] = y_pred
sample_submission.to_csv("submission.csv", index=False)
```

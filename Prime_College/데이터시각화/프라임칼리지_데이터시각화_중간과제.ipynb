```r
library(ggplot2)
load("/content/sah.RData")

library(RColorBrewer)

ggplot(sah, aes(BMI.cat, fill = BMI.cat)) +
  geom_bar() +
  scale_fill_brewer(palette = "Blues") +
  labs(
    title = "202457-352014",
    x = "BMI.cat",
    y = "Count",
    fill = "BMI.cat"
  )
```

```r
load("/content/sah.RData")

library(ggplot2)

ggplot(sah, aes(x = age)) +
  geom_histogram(
    breaks = seq(10, 65, by = 5),
    fill = "skyblue",
    color = "black"
  ) +
  labs(
    x = "Age in years",
    y = "Count",
    title = "202457-352014"
  )
```

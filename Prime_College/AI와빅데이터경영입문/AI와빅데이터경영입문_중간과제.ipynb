## seaborn 다이아몬드(diamonds) 데이터 분석

## (1) 데이터 로드하기

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import numpy as np

# 구글 코랩 한글 폰트 설정
import os
if 'COLAB_RELEASE_TAG' in os.environ:
    !apt-get install -y fonts-nanum > /dev/null 2>&1
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
    fm.fontManager.addfont(font_path)
    plt.rcParams['font.family'] = 'NanumGothic'
else:
    plt.rcParams['font.family'] = 'AppleGothic'

plt.rcParams['axes.unicode_minus'] = False

# 다이아몬드 데이터 로드
diamonds = sns.load_dataset('diamonds')

# 데이터 확인
print(f"데이터 크기: {diamonds.shape}")
print(f"\n컬럼 목록: {list(diamonds.columns)}")
diamonds.head()
```

```python
# 기본 통계량 확인
diamonds.describe()
```

```python
# 데이터 타입 및 결측치 확인
diamonds.info()
```

### 발견한 사실
- diamonds 데이터셋은 53,940개의 행과 10개의 열로 구성되어 있다.
- 수치형 변수: carat, depth, table, price, x, y, z (7개)
- 범주형 변수: cut, color, clarity (3개)
- 결측치는 존재하지 않는다.

## (2) 히스토그램(histplot) 그리기

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x="carat", kde=True)
plt.title('다이아몬드 캐럿(carat) 분포')
plt.xlabel('Carat')
plt.ylabel('Count')
plt.show()
```

### 해석
- 캐럿 분포는 오른쪽으로 긴 꼬리를 가진 양의 편포(right-skewed) 형태를 보인다.
- 대부분의 다이아몬드는 0.2~1.5 캐럿 사이에 집중되어 있다.
- 0.3캐럿과 1.0캐럿 부근에서 뚜렷한 봉우리가 관찰되며, 이는 소비자 선호도나 가격 구간의 영향일 수 있다.
- 2캐럿 이상의 대형 다이아몬드는 상대적으로 매우 드물다.

## (3) 상자그림(boxplot) 그리기

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="color", y="carat", hue="cut", data=diamonds, palette="pastel")
plt.title('색상(color)별 캐럿 분포 - 컷(cut) 기준')
plt.xlabel('Color')
plt.ylabel('Carat')
plt.show()
```

### 해석
- color가 D(무색에 가까움)에서 J(노란빛)로 갈수록 캐럿의 중앙값과 범위가 전반적으로 커지는 경향이 있다.
- 이는 큰 다이아몬드일수록 색상이 더 눈에 띄기 때문에 낮은 등급의 color로 분류되는 경우가 많기 때문으로 보인다.
- cut 등급별로 보면, Fair 등급의 다이아몬드가 상대적으로 캐럿이 큰 편이며, Ideal 등급은 캐럿이 작은 쪽에 몰려 있다.
- 각 그룹에서 이상치(outlier)가 상당수 관찰되며, 특히 고캐럿 구간에서 많이 나타난다.

## (4) 바이올린그림(violinplot) 그리기

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="cut", y="carat", data=diamonds, palette="pastel")
plt.title('컷(cut) 등급별 캐럿 분포')
plt.xlabel('Cut')
plt.ylabel('Carat')
plt.show()
```

### 해석
- Ideal과 Premium 등급은 주로 낮은 캐럿 구간에 데이터가 집중되어 있어 바이올린 모양이 아래쪽이 넓다.
- Fair 등급은 다른 등급에 비해 캐럿이 넓게 분포하고 있으며, 중앙값도 더 높은 편이다.
- 이는 캐럿이 큰 다이아몬드는 커팅 품질을 높이기 어렵기 때문에 Fair 등급에 많이 분포하는 것으로 해석할 수 있다.
- 모든 등급에서 꼬리가 위쪽으로 길게 뻗어 있어, 고캐럿 다이아몬드는 등급과 무관하게 소수 존재한다.

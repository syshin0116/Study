## seaborn 다이아몬드(diamonds) 데이터 분석

## (1) 데이터 로드하기

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import numpy as np

# 구글 코랩 한글 폰트 설정
import os
if 'COLAB_RELEASE_TAG' in os.environ:
    !apt-get install -y fonts-nanum > /dev/null 2>&1
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
    fm.fontManager.addfont(font_path)
    plt.rcParams['font.family'] = 'NanumGothic'
else:
    plt.rcParams['font.family'] = 'AppleGothic'

plt.rcParams['axes.unicode_minus'] = False

# 다이아몬드 데이터 로드
diamonds = sns.load_dataset('diamonds')

# 데이터 확인
print(f"데이터 크기: {diamonds.shape}")
print(f"\n컬럼 목록: {list(diamonds.columns)}")
diamonds.head()
```

```python
# 기본 통계량 확인
diamonds.describe()
```

```python
# 데이터 타입 및 결측치 확인
diamonds.info()
```

### 발견한 사실
- diamonds 데이터셋은 53,940개의 행과 10개의 열로 구성되어 있다.
- 수치형 변수: carat, depth, table, price, x, y, z (7개)
- 범주형 변수: cut, color, clarity (3개)
- 결측치는 존재하지 않는다.

## (2) 히스토그램(histplot) 그리기

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x="carat", kde=True)
plt.title('다이아몬드 캐럿(carat) 분포')
plt.xlabel('Carat')
plt.ylabel('Count')
plt.show()
```

### 해석
- 캐럿 분포는 오른쪽으로 긴 꼬리를 가진 양의 편포(right-skewed) 형태를 보인다.
- 대부분의 다이아몬드는 0.2~1.5 캐럿 사이에 집중되어 있다.
- 0.3캐럿과 1.0캐럿 부근에서 뚜렷한 봉우리가 관찰되며, 이는 소비자 선호도나 가격 구간의 영향일 수 있다.
- 2캐럿 이상의 대형 다이아몬드는 상대적으로 매우 드물다.

## (3) 상자그림(boxplot) 그리기

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="color", y="carat", hue="cut", data=diamonds, palette="pastel")
plt.title('색상(color)별 캐럿 분포 - 컷(cut) 기준')
plt.xlabel('Color')
plt.ylabel('Carat')
plt.show()
```

### 해석
- color가 D(무색에 가까움)에서 J(노란빛)로 갈수록 캐럿의 중앙값과 범위가 전반적으로 커지는 경향이 있다.
- 이는 큰 다이아몬드일수록 색상이 더 눈에 띄기 때문에 낮은 등급의 color로 분류되는 경우가 많기 때문으로 보인다.
- cut 등급별로 보면, Fair 등급의 다이아몬드가 상대적으로 캐럿이 큰 편이며, Ideal 등급은 캐럿이 작은 쪽에 몰려 있다.
- 각 그룹에서 이상치(outlier)가 상당수 관찰되며, 특히 고캐럿 구간에서 많이 나타난다.

## (4) 바이올린그림(violinplot) 그리기

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="cut", y="carat", data=diamonds, palette="pastel")
plt.title('컷(cut) 등급별 캐럿 분포')
plt.xlabel('Cut')
plt.ylabel('Carat')
plt.show()
```

### 해석
- Ideal과 Premium 등급은 주로 낮은 캐럿 구간에 데이터가 집중되어 있어 바이올린 모양이 아래쪽이 넓다.
- Fair 등급은 다른 등급에 비해 캐럿이 넓게 분포하고 있으며, 중앙값도 더 높은 편이다.
- 이는 캐럿이 큰 다이아몬드는 커팅 품질을 높이기 어렵기 때문에 Fair 등급에 많이 분포하는 것으로 해석할 수 있다.
- 모든 등급에서 꼬리가 위쪽으로 길게 뻗어 있어, 고캐럿 다이아몬드는 등급과 무관하게 소수 존재한다.

## (5) 상관계수 히트맵(heatmap) 그리기

```python
# 숫자형 변수만 추출
diamonds_num = diamonds.select_dtypes(include=['float64', 'int64'])

# 결측치 제거
diamonds_num_dropna = diamonds_num.dropna()

# 상관계수 행렬 계산
corr_matrix = diamonds_num_dropna.corr()

# 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(data=corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            square=True, linewidths=0.5)
plt.title('수치형 변수 간 상관계수 히트맵')
plt.tight_layout()
plt.show()
```

### 해석
- carat과 price 사이에 높은 양의 상관관계(약 0.92)가 관찰된다. 캐럿이 클수록 가격이 높아지는 경향이 뚜렷하다.
- x, y, z (다이아몬드의 물리적 크기)와 carat, price 간의 상관관계도 매우 높다(0.86~0.98). 크기가 크면 당연히 무겁고, 가격도 높아지기 때문이다.
- depth와 table은 다른 변수들과의 상관관계가 상대적으로 낮다. 이 두 변수는 다이아몬드의 비율에 관한 값이라 크기나 가격과는 직접적인 관련이 적다.
- x, y, z 간에는 거의 1에 가까운 상관관계가 있는데, 다이아몬드의 가로/세로/높이가 비례적으로 커지기 때문이다.

## 수행 과정에서 발견한 사실

1. diamonds 데이터셋은 총 53,940개의 다이아몬드 정보를 담고 있으며, 결측치가 없어 전처리 부담이 적었다.

2. 캐럿(carat)은 가격(price)과 가장 높은 상관관계를 보이는 변수로, 다이아몬드 가격을 결정하는 가장 주요한 요인임을 확인할 수 있었다.

3. cut 등급이 높다고(Ideal) 반드시 캐럿이 큰 것은 아니었다. 오히려 Fair 등급에서 캐럿이 큰 다이아몬드가 많이 관찰되었는데, 이는 큰 원석을 커팅할 때 품질을 유지하기 어렵기 때문으로 추측된다.

4. color 등급에서도 비슷한 패턴이 나타났다. 색상 등급이 낮은(J에 가까운) 다이아몬드가 평균적으로 더 큰 캐럿을 가지고 있었다.

5. x, y, z 세 변수는 서로 거의 완벽한 상관관계를 보여서, 분석 시 다중공선성 문제가 발생할 수 있다. 회귀분석을 한다면 이 중 하나만 사용하거나 차원축소를 고려해야 할 것이다.

## 분석 결과에 대한 해석

다이아몬드 데이터를 시각화한 결과, 가격에 가장 큰 영향을 미치는 요인은 캐럿(무게)이었다. 상관계수 히트맵에서 carat-price 간 상관계수가 0.92로 매우 높게 나타났으며, 물리적 크기(x, y, z)와도 높은 상관관계를 보였다.

흥미로운 점은 커팅 품질(cut)과 캐럿 사이의 관계이다. 바이올린 플롯과 박스플롯을 통해 확인한 결과, Ideal 등급의 다이아몬드는 대부분 1캐럿 이하의 작은 크기에 집중되어 있었고, Fair 등급은 상대적으로 넓은 캐럿 범위에 걸쳐 분포했다. 이는 큰 다이아몬드를 가공할 때 이상적인 커팅 비율을 유지하기가 더 어렵다는 점을 시사한다.

히스토그램을 보면 캐럿의 분포가 오른쪽으로 치우쳐 있어, 시장에서 거래되는 다이아몬드의 대다수가 소형이라는 것을 알 수 있었다. 0.3캐럿, 1.0캐럿 등 특정 구간에서 빈도가 급증하는 것은 소비자들이 선호하는 캐럿 구간이 있기 때문으로 보인다.

전체적으로 seaborn 라이브러리를 활용하면 비교적 간단한 코드로도 데이터의 분포와 변수 간 관계를 직관적으로 파악할 수 있다는 점을 실습을 통해 확인했다.
